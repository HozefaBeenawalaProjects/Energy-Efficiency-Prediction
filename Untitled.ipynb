```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import  RandomForestRegressor
from xgboost import XGBRFRegressor
```

```python
X = df[['Relative_compactness','Surface_area','Wall_area','Roof_area','Overall_height','Orientation','Glazing_area','Glazing_area_distribution']]
y = df[['Heating_load']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
pipeline_lr=Pipeline([('scalar1',StandardScaler()),
                     ('lr_regressor',LinearRegression())])
```

```python
pipeline_dt=Pipeline([('scalar2',StandardScaler()),
                     ('dt_regressor',DecisionTreeRegressor())])
```

```python
pipeline_rf=Pipeline([('scalar3',StandardScaler()),
                     ('dt_regressor',RandomForestRegressor())])
```

```python
pipeline_xg=Pipeline([('scalar4',StandardScaler()),
                     ('dt_regressor',XGBRFRegressor())])
```

```python
pipelines = [pipeline_lr, pipeline_dt, pipeline_rf, pipeline_xg]
```

```python
pipe_dict = {0: 'Linear Regression', 1: 'Decision Tree', 2: 'Random Forest', 3:'XG Boost'}
```

```python
for pipe in pipelines:
    pipe.fit(X_train, y_train)
```

```python
for i,model in enumerate(pipelines):
    print("{} r2 score: {}".format(pipe_dict[i],model.score(X_test,y_test)))
```

```python
best_accuracy=0.0
best_regressor=0
best_pipeline=""
for i,model in enumerate(pipelines):
    if model.score(X_test,y_test)>best_accuracy:
        best_accuracy=model.score(X_test,y_test)
        best_pipeline=model
        best_regressor=i
print('regressor with best accuracy:{}'.format(pipe_dict[best_regressor]))
```

```python
def predict_Heating_load (Relative_compactness, Surface_area, Wall_area, Roof_area, Overall_height, Orientation, Glazing_area, Glazing_area_distribution):
    
    x = np.zeros(len(X.columns))
    x[0] = Relative_compactness
    x[1] = Surface_area
    x[2] = Wall_area
    x[3] = Roof_area
    x[4] = Overall_height
    x[5] = Orientation
    x[6] = Glazing_area
    x[7] = Glazing_area_distribution
        
    return pipeline_rf.predict([x])[0]
```
